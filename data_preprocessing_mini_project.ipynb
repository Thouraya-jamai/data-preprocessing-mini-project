```python
import pandas as pd
pd.read_csv("pima-indians-diabetes.data.csv")
```

```python
columns_names =["NumTimesPrg","PlGlcConc","BloodP","SkinThick","TwoHourSerIns","BMI","DiPedFunc","age","HasDiabetes"]
data=pd.read_csv("pima-indians-diabetes.data.csv",names=columns_names)
data.head(10)
```

# data cleaning

```python
#1 . Filter the values of the 'SkinThick' attribute to determine non-zero values.
skinthick_non_null = data[data['SkinThick'] != 0]
print(skinthick_non_null.head())
```

```python
# 2. Calculate the median of these values
skinthik_median=data[data['SkinThick']!=0]['SkinThick'].median()
print("the median is: ",skinthik_median)
```

```python
#3. Replace zero values of the "SkinThick" attribute with the median.
data['SkinThick']=data['SkinThick'].replace(0,skinthik_median)
data.head(10)
```

```python
# 4. Why can't we do the same thing with the "NumTimesPrg" attribute?
In the case of the "NumTimesPrg" attribute (number of pregnancies), a zero could have a valid meaning, because a woman may not have had any pregnancies yet. In this case, replacing the zeros with the median might not be appropriate, as it might introduce incorrect information.
```
